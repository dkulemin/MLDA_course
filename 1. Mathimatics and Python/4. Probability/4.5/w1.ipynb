### Задача 1.1
Из ящика, содержащего три билета с номерами 1, 2, 3, вынимают по одному все билеты. Предполагается, что все последовательности номеров билетов имеют одинаковые вероятности. Найти вероятность того, что хотя бы у одного билета порядковый номер совпадает с собственным.

#### Подсказка
Постройте таблицу всех возможных решений (каким по порядку какой билет вынули)

#### Решение
Рассмотрим все возможные варианты (каждая ячейка это то каким по порядку можно вытащить каждый из билетов, 1 строка - 1 исход):

| 1й билет | 2й билет | 3й билет |
|----------|----------|----------|
| 1        | 2        | 3        |
| 1        | 3        | 2        |
| 3        | 1        | 2        |
| 3        | 2        | 1        |
| 2        | 1        | 3        |
| 2        | 3        | 1        |

Как видим всего возможно исходов 6 (Пространство элементарных исходов $|\Omega| = 6$)

Благоприятные исходы, это те строки где хотя бы одно из чисел в строке совпадает с номером столбца. Таких строк 4 (1, 2, 4, 5 => Пространство благоприятных элементарных исходов $|A| = 4$

Вероятность события А равна отношению числа благоприятных исходов ко всем возможным:

$P = \frac{|A|}{|\Omega|} = \frac{4}{6} = \frac{2}{3}$

#### Ответ:
$\frac{2}{3}$


#### Примечание
*Данную задачу так же можно рассматривать в рамках схемы Бернлли и полчить ответ без явного перебора исходов. См. неделю 2.*

### Задача 1.2

Колода из 36 карт хорошо перемешана (т. е. все возможные расположения карт равновероятны). Найти вероятности событий:

    а) А = {четыре туза расположены рядом},
    
    б) В = {места расположения тузов образуют арифметическую прогрессию с шагом 7}.
    
#### Подсказка 0
Задача сводится к комбинаторной на подсчёт количества перестановок n объектов
    
#### Подсказка 1а
При подсчёте числа возможных комбинаций представьте, что 4 туза представляют собой 1 "пачку" из 4 карт.

#### Подсказка 2а
Не забдьте, что разный порядок тузов внутри пачки приводит к разным комбинациям.

#### Подсказка 1б
Поставьте 1 туз на 1-е место и посмотрите какие места займут остальные тузы (при условии образования арифметической прогрессии). Сдвиньте 1й туз на 1 позицию. Посмотрите сколько всего может быть таких позиций.

#### Подсказка 2б
С учётом числа позиций тузов, занимаемые ими 4 места вновь можно считать "пачкой" как в пункте "а" и пересчитать по отдельности число перестановок внутри пачки и остальных чисел снаружи.


#### Решение

**а)** Число всех возможных исходов равно числу перестановок 36 карт, из курса комбинаторики мы знаем, что это 36!.

*Его можно рассчитать если представить что нам нужно разместить по 36 позициям в колоде 36 карт, тогда на 1-е свободное место мы можем поставить любую из 36 карт, на 2-е любую из 35 оставшихся (36-ю мы же использовали и разместить не можем), на 3-е любую из 34 (2 мы уже использовали) и т.д. пока не дойдём до последнего 36-ого места на которое сможем поставить всего 1 оставшуюся карту. Количество комбинаций, очевидно равно произведению количеству комбинаций для каждой позиции, т.е. $36\cdot 35\cdot 34\cdot 33\cdot...\cdot 1 = 36!$*

$|\Omega| = 36!$

Для рассчёта благоприятных событий представим, что все наши 4 туза идут "пачкой". Т.е. нам надо узнать число перестановок оставшихся 32 карт и 1 "пачки" по имеющимся 33 местам (32+1). Рассуждения аналогичны приведённым выше приводят к значению 33!

Однако учтём так же, что внутри "пачки" тузы могут быть расположены в любом порядке (и это будут разные, но всё ещё благопритные исходы). Таких перестановок будет 4! Суммарное же число перестановок благопритных событий будет равно произведению 33! и 4!

$|A| = 36!\cdot4!$


Вероятность события А равна отношению числа благоприятных исходов ко всем возможным:

$P = \frac{|A|}{|\Omega|} = \frac{33!4!}{36!} = \frac{4\cdot3\cdot2}{36\cdot35\cdot34} = \frac{1}{1785}$

   
 
    
    
**б)** Возможные позиции для тузов в арифметической прогрессии с шагом 7:

1 -> 8 -> 15 -> 22

...

15 -> 22 -> 29 -> 36

Т.е. мы можем 15-ю способами определить места, где будут находиться тузы.

Для каждого из этих способов остаётся 32! способов разместить остальные карты (32 оставшиеся карты по 32 позициям).

В свою очередь по 4 позициям 4 туза можно разместить 4! способами.

Т.е. число благоприятных исходов

$|B| = 15\cdot32!\cdot4!$

Число всех элементарных исходов не изменилось:

$|\Omega| = 36!$

Вероятность события B равна отношению числа благоприятных исходов ко всем возможным:

$P = \frac{|B|}{|\Omega|} = \frac{15\cdot32!\cdot4!}{36!} = \frac{15\cdot4\cdot3\cdot2}{36\cdot35\cdot34\cdot33} = \frac{15}{1785\cdot33} = \frac{1}{3927}$






#### Ответ:
а) $\frac{1}{1785}$

Б) $\frac{1}{3927}$

    а) А = {первая монета выпала «решкой» вверх},

$P = \frac{1}{2}$

### Задача 1.5

Из множества всех последовательностей длины n, состоящих из цифр 0, 1,2, случайно выбирается одна.

Найти вероятности событий:
    
- А = {последовательность начинается с 0},

- В = {последовательность содержит ровно m+2 нуля, причем 2 из них находятся на концах последовательности},
    
- С = {последовательность содержит ровно m единиц},
    
- D = {в последовательности ровно $m_{0}$ нулей, $m_{1}$ единиц, $m_{2}$ двоек}.

<span style="color: red;">**Важно** Задача требует знаний комбинаторики.</span>


#### Решение

Найдём число всех последовательностей длинны n из цифр 0, 1, 3. Очевидно, что на каждой позиции может находиться любая из них, а значит, это будет величина известная нам из курса комбинаторики как число размещений с повторениями из $\bar{A}_{3}^n = 3^n$.

$|\Omega| =3^n$

Т.к. событие А заключается в том, что на 1 позиции мы имеем цифру 0, то это означает, что оно состоит из всех последовательностей из 0 на 1-ом месте и любых других цир на последующих n-1 местах. Таких последовательностей $3^{n-1}$

$|A| =3^{n-1}$

Вероятность события А равна отношению числа благоприятных исходов ко всем возможным:

$P(A) = \frac{|A|}{|\Omega|}  = \frac{3^{n-1}}{3^n} = \frac{1}{3}$

------
#### Примечание 1
Для нахождения вероятности события А, как и в задачах 1.3 и 1.4а нам совершенно не важно что происходит с последовательностью **после** 1 элемента. Фактически нас интересует количество благоприятствующих состояний 1 символа и количество неблагоприятствующих. Очевидно, что раз благоприятсвует только число 1, а всего возможны 3 числа, то и вероятность такого события равна $\frac{1}{3}$

------

Для нахождения вероятности события B зафиксируем на 2 крайних местах нули. Теперь у нас осталось n-2 позиций.

Выберем из оставшихся n-2 позиций m-2 и разместим там нули. Из курса комбинаторики эта величина известна как число размещений без повторений: $C_{n-2}^{m} = \frac{(n-2)!}{(m)!(n-m-2)!}$

На оставшихся n-m-2 позициях мы можем разместить либо 1, либо 2 (т.к. по условию B содержит ровно m нулей и мы их все разместили). Следовательно это $2^{n-m-2}$.

$|B| =2^{n-m-2}\cdot C_{n-2}^{m}$


Вероятность события B равна отношению числа благоприятных исходов ко всем возможным:

$P(B) = \frac{|B|}{|\Omega|}  = \frac{2^{n-m-2}\cdot C_{n-2}^{m}}{3^n}$

Событие C даже проще, чем событие B. Выберем m позиций на которые поставим единицы. Это можно сделать $C_{n}^{m}$ способами.

На оставшиеся n-m позиций можно поставить либо 0, либо 2, а значит это можно сделать $2^{n-m}$ способами.

$|C| =2^{n-m}\cdot C_{n}^{m}$

$P(С) = \frac{|С|}{|\Omega|}  = \frac{2^{n-m}\cdot C_{n}^{m}}{3^n}$


Алгоритм нахождения вероятности события D аналогичен предыдущим двум, но будет происходить в 3 этапа для 0, 1 и 2:

- Выберем $m_{0}$ мест для нулей, это можно сделать $C_{n}^{m_{0}}$ способами.

- Выберем из $n-m_{0}$ оставшихся $m_{1}$ мест для единиц, это можно сделать $C_{n-m_{0}}^{m_{1}}$ способами.

- На оставшиеся $n-m_{0}-m_{1}$ местах разместим двойки. Обратим внимание, что это возможно только в случае, если $n-m_{0}-m_{1} = m_{2}$ и сделать это можно только 1 способом (все свободные места должны занять двойки).

$|D| =C_{n}^{m_{0}} \cdot C_{n-m_{0}}^{m_{1}} = \frac{n!}{m_{0}!(n-m_{0})!} \cdot \frac{(n-m_{0})!}{m_{1}!(n-m_{0}-m_{1})!} = \frac{n!}{m_{0}!m_{1}!(n-m_{0}-m_{1})!} = \frac{n!}{m_{0}!m_{1}!m_{2}!}$

------
#### Примечание 2
Это значение можно было сразу получить, используя формулу для полиномиального коэффициента $P(m_{0},m_{1},m_{2})$ (*не путать с P вероятностью!*)

------

$P(D) = \frac{|D|}{|\Omega|}  =\frac{n!}{m_{0}!m_{1}!m_{2}!} \cdot \frac{1}{3^n}$


#### Ответ:

$P(A) = \frac{1}{3}$

$P(B) = \frac{2^{n-m-2}\cdot C_{n-2}^{m}}{3^n}$

$P(С) = \frac{2^{n-m}\cdot C_{n}^{m}}{3^n}$

$P(D) =\frac{n!}{m_{0}!m_{1}!m_{2}!} \cdot \frac{1}{3^n}$


------
#### Примечание 3
Эту задачу логичнее и проще решать, используя схему Бернулли, см 2 неделю.

### Задача 1.6

Из 28 костей домино случайно выбираются две.
Найти вероятность $P_2$ того, что из них можно составить «цепочку» согласно правилам игры.

#### Подсказка 1

В интернете есть примеры решения "брутфорсом", т.е. с полным перебором всех благоприятных комбинаций. Интереснее рассмотреть задачу как выроятность извлечь 2 2-х символьных подмножества из множества {0,1,2,3,4,5,6} так, что внутри подмножества цифры могут повторяться, но сами подмножества не совпадают и циры из 1 подмножества не встречаются среди цифр 2.

#### Подсказка 2

РАссмотрите событие, обратное искомому - его вероятность найти проще.

#### Подсказка 3

Рассмотрите 3 случая по отдельности: извлечение 2 дублей, дубля и не дубля, и 2 не дублей.

#### Подсказка 4

Не забудьте, что при выборе дубля и не дубля важно определить какая из костей дубль, а какая нет. Это увеличивает количество возможных результатов.



#### Решение

Как найти число благоприятных исходов? В отличие от задачи про бросание монет тут очень тяжело построить таблицу возможных исходов. И хотя в интернете есть решения с похожим "брутфорсом" <noindex>(<a href="http://math.omsu.ru/info/learn/terver/0_1_3.htm" rel="nofollow">пример</a>)</noindex> мне они кажутся не очевидными. Поэтому я приведу иное решение.

Во-первых, представим кости домино как пары чисел, случайно извлекаемых из множества {0,1,2,3,4,5,6}.

В этом случае каждая кость домино - это процесс извлечения 2 чисел из этого множества с возможностью повторений.

$\overline{C}_{7}^{2} = C_{8}^{2} = \frac{8!}{6!\cdot 2!} = \frac{8\cdot 7}{2}=7\cdot 4=28$

И именно 28 костей домино в наборе! Т.е. извлечь 2 кости домино - это извлечь 2 пары чисел из множества {0,1,2,3,4,5,6} так, чтобы пары не совпадали, но числа внутри 1 пары могли совпадать. Выше приведено количество комбинаций для составления 1й пары (1я кость домино), по условию несовпадения пар чисел (костей домино), для второй пары остаётся на 1 комбинацию меньше. Всего же комбинаций:

$|\Omega| = 28\cdot27$

Рассмотрим теперь событие, обратное искомому, т.е. когда из двух костей нельзя составить цепочку. Какие у этого события могут быть варианты?

1) Мы извлекли 2 дубля (можно считать, что мы извлекли 1 число для 1й пары и 1 число из оставшихся для 2й пары). Число таких комбинаций определяется как $C_{7}^{1} \cdot C_{6}^{1}$

2) Мы извлекли дубль и НЕ дубль так, чтобы число на дубле не совпадало ни с одним из чисел на не дубле (сперва определем кто из пары будет дублем, потом выбираем для него 1 число из 7 возможных, а потом из 6 оставшихся выбираем 2 для недубля). Число комбинаций равно $2 \cdot C_{7}^{1} \cdot C_{6}^{2}$

3) Мы извлекли 2 НЕ дубля так, чтобы числа из 1 набора не встречались во 2. Число таких комбинаций определяется как $C_{7}^{2} \cdot C_{5}^{2}$

Число событий невозможности составить цепочку является суммой чисел всех 3 перечисленных комбинаций.

Вероятность выбрать 2 кости, которые НЕ составляют цепочтку, следовательно:

$P(не А) = \frac{C_{7}^{1} \cdot C_{6}^{1} + 2 \cdot C_{7}^{1} \cdot C_{6}^{2} + C_{7}^{2} \cdot C_{5}^{2}}{28\cdot 27} = \frac{\frac{7!6!}{6!1!5!1!} + 2\cdot \frac{7!6!}{6!1!4!2!} + \frac{7!5!}{5!2!3!2!}}{28\cdot 27} = \frac{\frac{7!}{5!} + 2\cdot \frac{7!}{2 \cdot 4!} + \frac{7!}{3!\cdot4}}{28\cdot 27} = \frac{7\cdot6 + 7\cdot 6 \cdot 5 + 7 \cdot 6 \cdot 5}{28\cdot 27} = \frac{7\cdot6 \cdot ( 1+ 5 + 5)}{7\cdot4 \cdot 3 \cdot 9} = \frac{11}{18}$


$P(A) = 1 - P(не А) = 1 - \frac{11}{18} = \frac{7}{18}$


#### Ответ:

$\frac{7}{18}$

### Задача 1.7

В записанном телефонном номере 135— 3.— .. три последние цифры стерлись. В предположении, что все комбинации трех стершихся цифр равновероятны, найти вероятности событий:

- А = {стерлись различные цифры, отличные от 1, 3, 5},

- В = {стерлись одинаковые цифры},

- С = {две из стершихся цифр совпадают}.


#### Решение

Событие А состоит в том, что на 3 последних позициях стоят цифры из множества {0, 2, 4, 6, 7, 8, 9}, причём без повторений. Т.е. число благопритных событий равно числу размещений без повторений из 7 по 3.

*Так же можно рассуждать, что на 1 из 3 пустых мест можно поставить одну из 7 цифр, на 2-е одну из 6 оставшихся, а на 3-е одну из 5 оставшихся.*

$|A| = A_{7}^{3} = 7 \cdot 6 \cdot 5$

Число всех возможных комбинаций:

$|\Omega| = 10^3$

$P(A) = \frac{|A|}{|\Omega|} = \frac{7 \cdot 6 \cdot 5}{1000} = 0.21 $

Если стёрлись одинаковые цифры, то это значит, что существет всего 10 благоприятных исходов (по числ возможных цифр).

$P(B) = \frac{|B|}{|\Omega|} = \frac{10}{1000} = 0.01 $

Для нахождения число комбинаций когда 2 цифры из 3 совпадают, выберем сперва 1 позицию где будет находиться несовпадающая с другими цифра (это можно сделать 3 способами), разместим на ней любю цифру (это можно сделать 10 способами), а на оставшиеся пустые места поставим две любых одинаковых цифры (это можно сделать 9 способами):

$|C| = 3 \cdot 10 \cdot 9$

$P(C) = \frac{|C|}{|\Omega|} = \frac{3 \cdot 10 \cdot 9}{1000} = 0.27 $

### Ответ:

- $P(A) = 0.21 $

- $P(B) = 0.01 $

- $P(C) = 0.27 $

### Задача 1.8

Какова вероятность того, что четырехзначный номер случайно взятого автомобиля в большом городе:

    а) имеет все цифры разные?
    
    б) имеет только две одинаковые цифры?
    
    в) имеет две пары одинаковых цифр?
    
    г) имеет только три одинаковые цифры?
    
    д) имеет все цифры одинаковые?



#### Подсказка

Воспользуйтесь знаниями <a rel="noindex" href ="https://www.coursera.org/learn/kombinatorika-dlya-nachinayushchikh/lecture/8uAgW/slova-v-iazykie">комбинаторики</a> (в первую очередь размещениями)


#### Решение

Число всех возможных 4 цифровых комбинаций из 10 возможных - это число размещений с повторениями:

$|\Omega| = \overline{A}_{10}^{4} = 10^4$

**a)** Число 4 цифровых комбинаций из 10 возможных - это число размещений без повторений:


$|A| =  A_{10}^{4} = 10 \cdot 9 \cdot 8 \cdot 7$

$P(A) = \frac{|A|}{|\Omega|} = \frac{10 \cdot 9 \cdot 8 \cdot 7}{10^4} = 0.504 $


**б)** Определим места для 2 одинаковых цифр из 4 мест, это будет число сочетаний без повторений из 4 по 2: $C_{4}^{2} = \frac{4!}{2!2!} = 6$

Теперь определим 3 числа, которые разместим (2 по 1 на место и 1 на 1 "двойное" место), это число размещений без повторений $A_{10}^{3} = 10 \cdot 9 \cdot 8$

$|B| = (10 \cdot 9 \cdot 8) \cdot 6$

$P(B) = \frac{|B|}{|\Omega|} = \frac{10 \cdot 9 \cdot 8 \cdot 6}{10^4} = 0.432 $


**в)** Выберем расположение пар. Это можно всего 3 способами (либо они идут подряд, либо через одного, либо одна пара находится "внутри" другой).

Теперь определим 2 числа, которые разместим, это число размещений без повторений $A_{10}^{2} = 10 \cdot 9$

$|C| = (10 \cdot 9) \cdot 3$

$P(C) = \frac{|C|}{|\Omega|} = \frac{10 \cdot 9 \cdot 3}{10^4} = 0.027 $

**г)** Выберем расположение 1 отлиной от 3 других цифр на 4 позициях, это будет число сочетаний без повторений из 4 по 1: $C_{4}^{1} = \frac{4!}{3!1!} = 4$

Теперь определим 2 числа, которые разместим, это число размещений без повторений $A_{10}^{2} = 10 \cdot 9$

$|D| = (10 \cdot 9) \cdot 4$

$P(D) = \frac{|D|}{|\Omega|} = \frac{10 \cdot 9 \cdot 4}{10^4} = 0.036 $

**д)** Выберем цифру, которая будет повторяться в благоприятных номерах 4 раза. Это можно сделать 10 способами.

$|E| = 10$

$P(E) = \frac{|E|}{|\Omega|} = \frac{10}{10^4} = 0.001 $


#### Ответ:

а) 0.504

б) 0.432

в) 0.027

г) 0.036

д) 0.001
