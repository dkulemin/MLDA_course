Давайте рассмотрим всех пользователей из контрольной группы (treatment = 1). Для таких пользователей мы хотим проверить гипотезу о том, что штат абонента не влияет на то, перестанет ли абонент пользоваться услугами оператора.

Для этого мы воспользуемся критерием хи-квадрат. Постройте таблицы сопряженности между каждой из всех 1275 возможных неупорядоченных пар штатов и значением признака churn. Для каждой такой таблицы 2x2 применить критерий хи-квадрат

```python
control = data[data.treatment == 1]
```

```python
control.head()
```

```python
unique_states = np.unique(control.state)
```

```python
unique_states
```

```python
state_pairs = []

for i, s1 in enumerate(unique_states):
    for j, s2 in enumerate(unique_states):
        if i >= j:
            continue
        
        state_pairs.append((s1, s2))
```

```python
%%time

tests = []

for pair in state_pairs:
    df = data[(data.treatment == 1) & (data.state.isin(pair))]
    table = pd.crosstab(df.state, df.churn)
    g, p, dof, expctd = stats.chi2_contingency(table, correction=False)
    tests.append([pair[0], pair[1], g, p])
    #print(table)
    #print(p)
    #print()
```

```python
state_churn_tests = pd.DataFrame.from_records(tests)
state_churn_tests.columns = ['state_A', 'state_B', 'statistic', 'p']
```

```python
(state_churn_tests.p < 0.05).value_counts()
```

```python
state_churn_tests.p.mean(), state_churn_tests.p.median()
```

```python
churn_0 = data[data.treatment == 0]['churn']
churn_2 = data[data.treatment == 2]['churn']
churn_cntrl = data[data.treatment == 1]['churn']
```

```python
churn_0 = churn_0.apply(lambda x: 1 if x == 'True.' else 0)
churn_2 = churn_2.apply(lambda x: 1 if x == 'True.' else 0)
churn_cntrl = churn_cntrl.apply(lambda x: 1 if x == 'True.' else 0)
```

```python
def proportions_confint_diff_ind(sample1, sample2, alpha = 0.05):    
    z = stats.norm.ppf(1 - alpha / 2.)   
    p1 = float(sum(sample1)) / len(sample1)
    p2 = float(sum(sample2)) / len(sample2)
    
    left_boundary = (p1 - p2) - z * np.sqrt(p1 * (1 - p1)/ len(sample1) + p2 * (1 - p2)/ len(sample2))
    right_boundary = (p1 - p2) + z * np.sqrt(p1 * (1 - p1)/ len(sample1) + p2 * (1 - p2)/ len(sample2))
    
    return (left_boundary, right_boundary)
```

```python
print("confidence interval for churn_0 - churn_cntrl: [%f, %f]" % proportions_confint_diff_ind(churn_0, churn_cntrl, alpha = 0.025))
```

```python
print("confidence interval for churn_2 - churn_cntrl: [%f, %f]" % proportions_confint_diff_ind(churn_2, churn_cntrl, alpha = 0.025))
```

```python
print("confidence interval for churn_2 - churn_0: [%f, %f]" % proportions_confint_diff_ind(churn_2, churn_0, alpha = 0.025))
```

```python
def proportions_diff_z_stat_ind(sample1, sample2):
    n1 = len(sample1)
    n2 = len(sample2)
    
    p1 = float(sum(sample1)) / n1
    p2 = float(sum(sample2)) / n2 
    P = float(p1*n1 + p2*n2) / (n1 + n2)
    
    return (p1 - p2) / np.sqrt(P * (1 - P) * (1. / n1 + 1. / n2))
```

```python
def proportions_diff_z_test(z_stat, alternative = 'two-sided'):
    if alternative not in ('two-sided', 'less', 'greater'):
        raise ValueError("alternative not recognized\n"
                         "should be 'two-sided', 'less' or 'greater'")
    
    if alternative == 'two-sided':
        return 2 * (1 - stats.norm.cdf(np.abs(z_stat)))
    
    if alternative == 'less':
        return stats.norm.cdf(z_stat)

    if alternative == 'greater':
        return 1 - stats.norm.cdf(z_stat)
```

```python
print(f"p-value: {proportions_diff_z_test(proportions_diff_z_stat_ind(churn_0, churn_cntrl))}")
```

```python
print(f"p-value: {proportions_diff_z_test(proportions_diff_z_stat_ind(churn_2, churn_cntrl))}")
```

```python
print(f"p-value: {proportions_diff_z_test(proportions_diff_z_stat_ind(churn_2, churn_0))}")
```
